```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Analysis of Super Store - DA.csv")
```

```python
#Sales and Profit analysis based on region
fig = plt.figure() # create figure
ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot)
colors = ['orange','green','yellow','blue']

df.groupby('Region')['Sales'].sum().plot.bar(figsize=(13,4),color=colors,ax=ax0)
df.groupby('Region')['Profit'].sum().plot.bar(figsize=(13,4),color=colors,ax=ax1)

ax0.set_xlabel('Region')
ax0.set_ylabel('Sales')
ax1.set_xlabel('Region')
ax1.set_ylabel('Sales')
plt.suptitle('Sales and Profit by region')
```

```python
# Sales and Profit Percentage Based on Segment
fig = plt.figure() # create figure
ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot)

df.groupby('Segment')['Sales'].sum().plot.pie(autopct = "%1.0f%%",figsize = (10,4),ax = ax0)
df.groupby('Segment')['Profit'].sum().plot.pie(autopct = "%1.0f%%",figsize = (10,4),ax = ax1)
plt.suptitle('Sales and Profit Percentage Based on Segment')
ax0.legend(loc='upper left') 
```

```python
# Sales and Profit Percentage Based on Category of the products 
fig = plt.figure() # create figure
ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot)

df.groupby('Category')['Sales'].sum().plot.pie(autopct = "%1.0f%%",figsize = (10,4),ax = ax0)
df.groupby('Category')['Profit'].sum().plot.pie(autopct = "%1.0f%%",figsize = (10,4),ax = ax1)
ax0.legend(loc='upper left') 
plt.suptitle('Sales and Profit Percentage Based on Category of the products')
```

```python
# Visualizing Profits according to States
df.groupby('State')['Profit'].sum().plot.bar(figsize=(12,5))
plt.xlabel('State')
plt.ylabel('Profit')
plt.title('Profits by States')
```

```python
#Visualizng Loss making States
neg_profit = df[df['Profit'] < 0]
neg_profit.groupby('State')['Profit'].sum().plot.bar(figsize=(8,4))
plt.ylim(0,-40000)
plt.xlabel('State')
plt.ylabel('Profit')
plt.title("Loss Making States")
```
